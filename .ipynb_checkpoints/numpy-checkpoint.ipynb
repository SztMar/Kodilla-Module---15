```python
import pandas as pd
import matplotlib.pyplot as plt

prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]

```

```python
df = pd.DataFrame(prices, columns=["month","price"])
df["priceUSD"] = df["price"]/4
df = df.set_index("month")


fig, ax = plt.subplots()
line, = ax.plot(df.index, df["priceUSD"], label='Using set_dashes()')
line.set_dashes([4, 2, 4, 2])
line.set_color('red')
line.set_linewidth(3)
ax.set_ylim(0,2)
ax.set_xlabel("month")
ax.set_ylabel("price")
ax.set_title("Price of goods (USD)")

ax.grid()
plt.show()
```
